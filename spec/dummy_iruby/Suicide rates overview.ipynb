## Finding the total female, male suicide count in different years and the country with highest suicide rates

**This notebook uses a 'Suicide Rates Overview 1985 to 2016' dataset from kaggle.**

```ruby
require 'daru'
```

```ruby
data_frame = Daru::DataFrame.from_csv("../dummy_data/suicides_data.csv")
data_frame
```

**Pivot the DataFrame on the year as the row and sex as the column, using the 'suicides_no' column for aggregation.**

**This tells us the number of male and female suicides per year.**

```ruby
data_frame.pivot_table(index: ['year'], values: 'suicides_no', vectors: ['sex'], agg: :sum)
```

**Accessing columns country and suicides_no**

```ruby
countries_df = data_frame['country', 'suicides_no']
```

**Pivot the DataFrame on the country as the row using the 'suicides_no' column for aggregation.**

```ruby
grouped_df = countries_df.pivot_table(index: ['country'], values: 'suicides_no', agg: :sum)
```

**Country with maximum suicides**

```ruby
grouped_df['suicides_no'].max_index
```
